```python
import pickle

def dump(fn, thing):
  with open(fn, "wb") as f:
    pickle.dump(thing, f)
def load(fn):
  with open(fn, "rb") as f:
    return pickle.load(f)
```

```python
def myplot(x, k):
    y = []
    y0 = []
    y1 = []
    for a in x:
        print(a)
        d = finalexam(labels, vectors, get_knn(a, k))
        y.append(d['Accuracy in average'])
        y0.append(d['Accuracy on falses'])
        y1.append(d['Accuracy on trues'])
    plt.plot(x, y, label='Accuracy in average')
    plt.plot(x, y0, label='Accuracy on falses')
    plt.plot(x, y1, label='Accuracy on trues')

    plt.xlabel('balance value')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)
    plt.ylim(0.1, 1)
    plt.show()
```

```python
myplot(list(range(1, 5)), 5)
```
